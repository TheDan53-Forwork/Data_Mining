##  1.  Cài đặt thuật toán Apriori

```python
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder


PATH = "data.csv"

# Load dataset
df =pd.read_csv(PATH)
display(df)
```

```python
# Convert data frame to list
records = []
for i in range(0, df.shape[0]):
    records.append([str(df.values[i,j]) for j in range(0, df.shape[1])])
```

```python
records
```

```python
# Convert records to transaction
te = TransactionEncoder()
te_ary = te.fit(records).transform(records)
dfl = pd.DataFrame(te_ary, columns=te.columns_)

display(dfl)
```

```python
frequent_itemsets = apriori(dfl, min_support=0.6, use_colnames=True)
display(frequent_itemsets)
```

```python
# Build association rules using support metric
rules = association_rules(frequent_itemsets, metric = 'support', support_only=True, min_threshold=0.1)

rule = rules[['antecedents', 'consequents', 'support']]
display(rule)
```

##  2.  Cài đặt lại thuật toán Apritori

```python
class MyTransactionEncoder:
    def __init__(self):
        self.unique_items = set()
        self.column_names = []  # Lưu trữ tên của các cột
        
    def fit(self, transactions):
        for transaction in transactions:
            for item in transaction:
                self.unique_items.add(item)
        self.unique_items = sorted(self.unique_items)
        self.column_names = self.unique_items  # Lưu tên cột dưới dạng các mặt hàng được sắp xếp
        return self
        
    def transform(self, transactions):
        item_index = {item: idx for idx, item in enumerate(self.unique_items)}
        transaction_matrix = []
        for transaction in transactions:
            row = [False] * len(self.unique_items)
            for item in transaction:
                if item in item_index:
                    row[item_index[item]] = True
            transaction_matrix.append(row)
        return transaction_matrix
        
    def fit_transform(self, transactions):
        self.fit(transactions)
        return self.transform(transactions)
    
    def get_feature_names(self):
        """ Trả về danh sách các tên cột, tương ứng với các mặt hàng duy nhất """
        return self.column_names
```

```python
myrecords = records.copy()

encoder = MyTransactionEncoder()
myEncodeArray = encoder.fit_transform(myrecords)

mydfl = pd.DataFrame(myEncodeArray, columns=encoder.get_feature_names())

display(mydfl)
```

```python
def MyApriori(transactions, min_support_ratio):
    import itertools
    
    # Bước 1: Tính tần số xuất hiện của từng mục trong tất cả giao dịch
    item_counts = {}
    for transaction in transactions:
        for item in transaction:
            if item in item_counts:
                item_counts[item] += 1
            else:
                item_counts[item] = 1  
    display(item_counts)     
    num_transactions = len(transactions)         
    min_support_count = int(min_support_ratio * num_transactions)
    
    # Lọc ra những mục có độ hỗ trợ đủ lớn
    items = {item for item, count in item_counts.items() if count >= min_support_count}
    
    # Tạo tập hợp thường xuyên cho từng kích thước từ 1
    freq_sets = []
    current_sets = [{item} for item in items]
    while current_sets:
        # Tạo tập hợp ứng viên mới từ các tập hiện tại
        candidate_sets = [x.union(y) for x in current_sets for y in current_sets if len(x.union(y)) == len(x) + 1]
        candidate_sets = list(set(map(frozenset, candidate_sets)))  # Loại bỏ trùng lặp
        
        # Tính hỗ trợ cho mỗi tập ứng viên
        valid_sets = []
        candidate_counts = {cand: 0 for cand in candidate_sets}
        for transaction in transactions:
            transaction_set = set(transaction)
            for candidate in candidate_sets:
                if candidate.issubset(transaction_set):
                    candidate_counts[candidate] += 1
        
        # Lọc ra những tập có hỗ trợ đủ lớn
        for candidate, count in candidate_counts.items():
            if count >= min_support:
                valid_sets.append(candidate)
        
        freq_sets.extend(valid_sets)
        current_sets = valid_sets
    
    return freq_sets

# Ví dụ sử dụng
min_support_ratio = 0.6
display(MyApriori(myrecords, min_support_ratio))
```
