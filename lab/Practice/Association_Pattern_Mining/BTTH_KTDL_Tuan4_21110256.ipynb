##  1.  Cài đặt thuật toán Apriori

```python
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder


PATH = "data.csv"

# Load dataset
df =pd.read_csv(PATH)
display(df)
```

```python
# Convert data frame to list
records = []
for i in range(0, df.shape[0]):
    records.append([str(df.values[i,j]) for j in range(0, df.shape[1])])
```

```python
records
```

```python
# Convert records to transaction
te = TransactionEncoder()
te_ary = te.fit(records).transform(records)
dfl = pd.DataFrame(te_ary, columns=te.columns_)

display(dfl)
```

```python
frequent_itemsets = apriori(dfl, min_support=0.6, use_colnames=True)
display(frequent_itemsets)
```

```python
# Build association rules using support metric
rules = association_rules(frequent_itemsets, metric = 'support', support_only=True, min_threshold=0.1)

rule = rules[['antecedents', 'consequents', 'support']]
display(rule)
```

##  2.  Cài đặt lại thuật toán Apritori

```python
class MyTransactionEncoder:
    def __init__(self):
        self.unique_items = set()
        self.column_names = []  # Lưu trữ tên của các cột
        
    def fit(self, transactions):
        for transaction in transactions:
            for item in transaction:
                self.unique_items.add(item)
        self.unique_items = sorted(self.unique_items)
        self.column_names = self.unique_items  # Lưu tên cột dưới dạng các mặt hàng được sắp xếp
        return self
        
    def transform(self, transactions):
        item_index = {item: idx for idx, item in enumerate(self.unique_items)}
        transaction_matrix = []
        for transaction in transactions:
            row = [False] * len(self.unique_items)
            for item in transaction:
                if item in item_index:
                    row[item_index[item]] = True
            transaction_matrix.append(row)
        return transaction_matrix
        
    def fit_transform(self, transactions):
        self.fit(transactions)
        return self.transform(transactions)
    
    def get_feature_names(self):
        """ Trả về danh sách các tên cột, tương ứng với các mặt hàng duy nhất """
        return self.column_names
```

```python
myRecords = records.copy()

encoder = MyTransactionEncoder()
myEncodeArray = encoder.fit_transform(myRecords)

my_dfl = pd.DataFrame(myEncodeArray, columns=encoder.get_feature_names())

display(my_dfl)
```

```python
def MyApriori(transactions, min_support):
    from itertools import combinations
    
    # Hàm này sẽ trả về tất cả các mục duy nhất trong tất cả các giao dịch
    def items_in_transactions(transactions):
        items = set()
        for transaction in transactions:
            items.update(transaction)
        return items
    
    # Hàm này tạo ra các tập ứng viên k-item từ các tập thường xuyên (k-1)-item
    def generate_candidate_sets(frequent_sets, k):
        return set([
            frozenset(set1 | set2)
            for set1 in frequent_sets
            for set2 in frequent_sets
            if len(set1 | set2) == k
        ])
    
    # Hàm này tính tỷ lệ hỗ trợ cho một tập ứng viên cụ thể
    def calculate_support(candidate_set, transactions):
        return sum(1 for transaction in transactions if candidate_set.issubset(transaction)) / len(transactions)
    
    # Lấy tất cả các mục từ các giao dịch
    items = items_in_transactions(transactions)
    # Khởi tạo các tập ứng viên 1-item
    candidate_1_item_sets = [{item} for item in items]
    frequent_sets = []
    support_sets = []
    
    # Xác định các tập thường xuyên 1-item
    for candidate in candidate_1_item_sets:
        support = calculate_support(candidate, transactions)
        if support >= min_support:
            frequent_sets.append(frozenset(candidate))
            support_sets.append(support)
    
    # Khởi tạo k = 2
    k = 2
    while True:
        # Tạo các tập ứng viên k-item từ các tập thường xuyên (k-1)-item
        candidate_k_item_sets = generate_candidate_sets(frequent_sets, k)
        current_frequent_sets = []
        current_support_sets = []
        for candidate_set in candidate_k_item_sets:
            support = calculate_support(candidate_set, transactions)
            if support >= min_support:
                current_frequent_sets.append(candidate_set)
                current_support_sets.append(support)
        
        # Nếu không còn tập ứng viên mới thường xuyên, quá trình sẽ dừng lại
        if not current_frequent_sets:
            break
        
        # Cập nhật các tập thường xuyên và tăng k lên 1
        frequent_sets.extend(current_frequent_sets)
        support_sets.extend(current_support_sets)
        k += 1
    
    # Trả về tất cả các tập thường xuyên tìm được
    return frequent_sets, support_sets
```

```python
frequent_itemsets = MyApriori(myRecords, min_support=0.6)
display(pd.DataFrame(frequent_itemsets))
```
